## Log Transformation for all variables

```python
log_candidates = [
    'AnnualIncome',
    'MonthlyIncome',
    'LoanAmount',
    'MonthlyLoanPayment',
    'MonthlyDebtPayments',
    'SavingsAccountBalance',
    'CheckingAccountBalance',
    'TotalAssets',
    'TotalLiabilities',
    'NetWorth'
]
```

```python
conditional = [
    'NumberOfOpenCreditLines',
    'NumberOfCreditInquiries',
    'LoanDuration'
]
```

```python
skewed_features = (
    df[log_candidates + conditional]
    .skew()
    .sort_values(ascending=False)
)

print(skewed_features)
```

```python
# List to be transformed
log_transform = [
    'SavingsAccountBalance',
    'TotalLiabilities',
    'NetWorth',
    'TotalAssets',
    'CheckingAccountBalance',
    'MonthlyLoanPayment',
    'AnnualIncome',
    'MonthlyIncome',
    'LoanAmount',
    'MonthlyDebtPayments'
]
```

```python
# Check - Apply Transformation if 0, use log1p
(df[log_transform] <=0).sum()
```

```python
# Apply Transformation
for col in log_transform:
    df[f'{col}_log'] = np.log1p(df[col])
```

```python
# Drop the old variables
df = df.drop(columns = log_transform)
```

```python
# Validate the fix
df[[f'{c}_log' for c in log_transform]].skew().sort_values(ascending=False)
```

## Two level modelling - DomainRiskScore & ML model

```python
# Domain Risk Score
# 1. Debt-to-income Risk
df['DTI_norm'] = np.minimum(df['DebtToIncomeRatio'] / 0.6, 1.0)

# 2. Credit utilization Risk
df['Util_norm'] = np.minimum(df['CreditCardUtilizationRate'] / 0.9, 1.0)

# 3. Loan-to-income Risk
df['LoanToIncome'] = df['LoanAmount_log'] / df['AnnualIncome_log']
df['LTI_norm'] = np.minimum(df['LoanToIncome'] / 1.0, 1.0)

# 4. Employment Risk
employment_risk_map = {
    'Unemployed': 1.0,
    'Self-employed': 0.6,
    'Contract': 0.5,
    'Part-time': 0.4,
    'Full-time': 0.2
}
df['EmploymentRisk'] = df['EmploymentStatus'].map(employment_risk_map)

# Together
df['DomainRiskScore'] = (
    0.35 * df['DTI_norm'] +
    0.25 * df['Util_norm'] +
    0.25 * df['LTI_norm'] +
    0.15 * df['EmploymentRisk']
)
```
