```python
# Apply log1p - Log Transformation

df['MonthlyIncome_log'] = np.log1p(df['MonthlyIncome'])
df['AnnualIncome_log'] = np.log1p(df['AnnualIncome'])
```

```python
# Validate the transformation
print("Original skew:", df['MonthlyIncome'].skew())
print("Log skew:", df['MonthlyIncome_log'].skew())

# Annual
print("Original Annual skew:", df['AnnualIncome'].skew())
print("Log Annual skew:", df['AnnualIncome_log'].skew())
```

### Income (Annual & Monthly) exhibits strong positive skewness, indicating a small number of high-income applicants. Log transformation was applied to stabilize variance and improve regression performance.

## Log Transformation for all variables

```python
log_candidates = [
    'AnnualIncome',
    'MonthlyIncome',
    'LoanAmount',
    'MonthlyLoanPayment',
    'MonthlyDebtPayments',
    'SavingsAccountBalance',
    'CheckingAccountBalance',
    'TotalAssets',
    'TotalLiabilities',
    'NetWorth'
]
```

```python
conditional = [
    'NumberOfOpenCreditLines',
    'NumberOfCreditInquiries',
    'LoanDuration'
]
```

```python
skewed_features = (
    df[log_candidates + conditional]
    .skew()
    .sort_values(ascending=False)
)

print(skewed_features)
```

```python
# List to be transformed
log_transform = [
    'SavingsAccountBalance',
    'TotalLiabilities',
    'NetWorth',
    'TotalAssets',
    'CheckingAccountBalance',
    'MonthlyLoanPayment',
    'AnnualIncome',
    'MonthlyIncome',
    'LoanAmount',
    'MonthlyDebtPayments'
]
```

```python
# Check - Apply Transformation if 0, use log1p
(df[log_transform] <=0).sum()
```

```python
# Apply Transformation
for col in log_transform:
    df[f'{col}_log'] = np.log1p(df[col])
```

```python
# Drop the old variables
df = df.drop(columns = log_transform)
```

```python
# Validate the fix
df[[f'{c}_log' for c in log_transform]].skew().sort_values(ascending=False)
```
